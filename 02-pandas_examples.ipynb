```python
df['Births'].plot()

filter01 = df['Births'] == df['Births'].max()
max_name = df['Names'][filter01].values

max_value = df['Births'].max()
anno_text = str(max_name) +  '-' + str(max_value)

max_value_index = df.sort_values(['Births'],ascending=False).head(1).index[0]
plt.scatter(max_value_index,max_value,marker='o',color='orange',s=45)
plt.annotate(1,1,anno_text)
```
